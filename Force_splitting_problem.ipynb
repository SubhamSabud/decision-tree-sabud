```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a dataset
# Sample data with two features (x1 and x2)
np.random.seed(42)

# Class 0 data points (blue)
class_0_points = np.array([[2, 2], [3, 3], [4, 4], [5, 5]])

# Class 1 data points (red)
class_1_points = np.array([[7, 7], [8, 8], [9, 9], [10, 10]])

# Combine the data points
X = np.vstack((class_0_points, class_1_points))
y = np.hstack((np.zeros(len(class_0_points)), np.ones(len(class_1_points))))

# Step 2: Plot the dataset
plt.scatter(class_0_points[:, 0], class_0_points[:, 1], c='blue', label='Class 0')
plt.scatter(class_1_points[:, 0], class_1_points[:, 1], c='red', label='Class 1')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Training Set')
plt.grid(True)
plt.show()

# Step 3: Manually force a split
# If we split at x1 = 6, the algorithm can create a deeper tree with zero training error
plt.scatter(class_0_points[:, 0], class_0_points[:, 1], c='blue', label='Class 0')
plt.scatter(class_1_points[:, 0], class_1_points[:, 1], c='red', label='Class 1')
plt.axvline(x=6, color='green', linestyle='--', label='Manual Split at x1 = 6')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Training Set with Manual Split')
plt.grid(True)
plt.show()
```
